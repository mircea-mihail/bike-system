```python
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
def show_pic(img):
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

```python
img_to_convert = "give-way.png"
img = cv.imread(img_to_convert)
hsv_img = cv.cvtColor(img, cv.COLOR_BGR2HSV)

w = img.shape[1]
h = img.shape[0]
show_pic(img)

for i in range(h):
    for j in range(w):
        if((hsv_img[i][j][0] < 10 or hsv_img[i][j][0] > 230) and hsv_img[i][j][1] > 190 and hsv_img[i][j][2] > 190):
            if (i + j) % 2 == 0:
                img[i, j] = (0, 0, 0)      
            else:
                img[i, j] = (255, 255, 255)

show_pic(img)
cv.imwrite("checker-" + img_to_convert, img)
```
